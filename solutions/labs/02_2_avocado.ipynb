```python
# https://www.kaggle.com/neuromusic/avocado-prices
import pandas as pd
```

## Tasks

```python
# 1
data = pd.read_csv('avocado.csv')
```

```python
# 7
data.AveragePrice
```

```python
# 8
data['EstimatedRevenue'] = data['Total Volume'] * data['AveragePrice']
data.head()
```

```python
# 9
data_grouped = data.groupby(['region','type'])[['AveragePrice']].mean()
data_grouped.reset_index(inplace=True)
data_grouped.head()
```

```python
# 10
data.groupby('year')['Total Volume'].agg(['mean','median','std']).plot.bar()
```
